# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `MNIST handwritten digits` (arquivos de treino e teste)
Realize a separação do conjunto de treino como treino e validação
Utilize `train_test_split` com controle de aleatoriedade (seed)
Retorne: `X_train`, `X_val`, `y_train`, `y_val`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

```python
import pandas as pd
from sklearn.model_selection import train_test_split


def load_data(seed):
    train = pd.read_csv("mnist_train.csv")
    X = train.drop(["label"])
    y = train["label"]

    X_train, X_val, y_train, y_val = train_test_split(
        X, y, test_size=0.2, random_state=seed
    )

    return X_train, X_val, y_train, y_val
```

Não é necessário normalizar os dados para esse tipo de modelo, porque modelos baseados em árvores não dependem da escala das variáveis da mesma forma que modelos baseados em distância ou gradiente. Eles tomam decisões por meio de divisões nos atributos, então mudar a escala dos pixels geralmente não altera de forma relevante o funcionamento do modelo.

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

```python
from sklearn.ensemble import RandomForestClassifier

def train_random_forest(X_train, y_train, seed):
    model = RandomForestClassifier(random_state=seed)
    model.fit(X_train, y_train)
    return model
```

# Questão 3

Implemente a função:

- `evaluate(model, X_test, y_test)`

Ela deve:
- Realizar predições
- Retornar a acurácia do modelo

**Solução**:

```python
from sklearn.metrics import accuracy_score

def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    return acc
```

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

```python
def run_pipeline(model_type = "rf", seed = 42):
    X_train, X_val, y_train, y_val = load_data(seed)

    if model_type = "rf":
        model = train_random_forest(X_train, y_train, seed)

    acc = evaluate(model, X_val, y_val)

    return acc
```

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O overfitting começa na profundidade em que o desempenho no treino melhora, mas o desempenho na validação piora ou deixa de melhorar. Isso mostra que o modelo está aprendendo demais os detalhes do treino.A árvore atinge 100% no treino com `max_depth=None` porque não há limite para as divisões. Assim, ela pode memorizar os dados de treino em vez de aprender padrões mais gerais.

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest

## Apresente:
- Acurácia, Precisão, Recall e F1-Score para o modelo

**Solução**:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

X_train, X_val, y_train, y_val = load_data(42)
model = train_random_forest(X_train, y_train, 42)

y_pred = model.predict(X_val)

acc = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred, average="weighted")
recall = recall_score(y_val, y_pred, average="weighted")
f1 = f1_score(y_val, y_pred, average="weighted")

print("Acurácia:", acc)
print("Precisão:", precision)
print("Recall:", recall)
print("F1-Score:", f1)


```

# Questão 7

Para pelo menos um dos modelos:

- Compare a acurácia em treino e teste

## Responda:
- Existe overfitting?
- Qual modelo tende a sofrer mais com isso?

```python
X_train, X_val, y_train, y_val = load_data(42)

model = train_random_forest(X_train, y_train, 42)

acc_train = evaluate(model, X_train, y_train)
acc_val = evaluate(model, X_val, y_val)

print("Acurácia no treino:", acc_train)
print("Acurácia na validação:", acc_val)
```

Há indícios de overfitting quando a acurácia no treino é maior do que a acurácia no teste/validação, pois isso mostra que o modelo aprendeu muito bem os dados vistos no treinamento, mas teve desempenho inferior em dados novos.

Entre os modelos, a árvore de decisão tende a sofrer mais com overfitting, porque ela pode crescer demais e memorizar os exemplos de treino. Já o Random Forest costuma ser mais robusto, pois combina várias árvores e reduz esse efeito.

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`

## Analise:
- O desempenho muda significativamente?

```python
from sklearn.ensemble import RandomForestClassifier

X_train, X_val, y_train, y_val = load_data(42)

valores = [10, 50, 100, 200]

for n in valores:
    model = RandomForestClassifier(
        n_estimators=n,
        random_state=42
    )
    
    model.fit(X_train, y_train)
    acc = evaluate(model, X_val, y_val)
    
    print(f"n_estimators={n} -> acurácia={acc}")
```

O desempenho pode mudar ao variar `n_estimators`, mas normalmente essa mudança não é muito grande após certo ponto. Em geral, aumentar o número de árvores tende a deixar o modelo mais estável e, às vezes, melhorar um pouco a acurácia, mas os ganhos costumam diminuir conforme o valor cresce.

Assim, o impacto existe, porém nem sempre é significativo. Muitas vezes, depois de uma certa quantidade de árvores, o custo computacional aumenta mais do que a melhoria no desempenho.
